```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Heat sink with 12 fins

```python
# Read the CSV file, skipping the first 4 rows and specifying the delimiter
file_path = "Thermal_12_data.csv"
data = pd.read_csv(file_path, skiprows=4)

# Check the exact column names
print(data)
print(data.columns)

# Extract the data
X1 = data["X (mm)  "]
Y1 = data["Y (mm)  "]
Z1 = data["Z (mm)  "]
T1 = data["TEMP (Celsius)"]
nodes = data["Node  "]

# Scatter X versus temperature
plt.scatter(X1, T1, c = "red", marker = "o", s = 5)

# Customize the plot 
plt.xlabel('X (mm)')
plt.ylabel('T (Celsius)')
plt.title('Scatter Plot of X vs Temperature for Heat Sink with 12 Fins', fontsize = 9)


# Check if the final temperature matches the scattered maximum temperature
max_temperature = data["TEMP (Celsius)"].max()
print(f"The maximum temperature is: {max_temperature}")


# Save the plot
plt.savefig('Temperature_vs_X_12_fins', dpi=300)  

# Display the plot
plt.show()
```

```python
# Scatter Y versus temperature
plt.scatter(Y1, T1, c = "red", marker = "o", s = 5)

# Customize the plot 
plt.xlabel('Y (mm)')
plt.ylabel('T (Celsius)')
plt.title('Scatter Plot of Y vs Temperature for Heat Sink with 12 Fins', fontsize = 9)

# Save and show the plot
plt.savefig('Temperature_vs_Y_12_fins', dpi=300)  # You can specify the file format with the extension (e.g., .png, .jpg, .pdf)

plt.show()
```

```python
# Scatter Z versus temperature
plt.scatter(Z1, T1, c = "red", marker = "o", s = 5)

# Step 4: Customize the plot 
plt.xlabel('Z (mm)')
plt.ylabel('T (Celsius)')
plt.title('Scatter Plot of Z vs Temperature for Heat Sink with 12 Fins', fontsize = 9)

plt.savefig("Temperature_vs_Z_12_fins", dpi=300)

plt.show()
```
